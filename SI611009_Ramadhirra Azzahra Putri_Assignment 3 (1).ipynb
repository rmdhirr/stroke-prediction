# **Background**
Stroke represents a major global health issue, with its high rates of mortality and severe long-term disabilities significantly impacting public health systems worldwide. According to the World Health Organization (WHO), strokes are the second leading cause of death globally, accounting for approximately 6 million deaths each year. Moreover, strokes are a leading cause of disability, leaving many survivors with significant cognitive and physical impairments that require long-term care. The burden of stroke is not only a medical challenge but also a significant economic strain on families and healthcare systems.

Effective stroke prevention hinges on the accurate identification of individuals at high risk based on modifiable and non-modifiable risk factors such as age, hypertension, diabetes, and lifestyle choices. Traditional epidemiological approaches have utilized statistical methods to identify risk factors for stroke, but these methods often fail to capture complex interactions between variables and might not adequately model the nonlinear dynamics of stroke risk.

# **Problem Statement**
Despite extensive research and numerous advancements in medical science, the predictive models currently used in stroke prevention have limitations in accuracy, adaptability, and scalability. Conventional statistical models often cannot fully integrate and analyze the multifaceted and interactive effects of risk factors associated with strokes. This limitation leads to inefficiencies in identifying at-risk individuals, potentially resulting in delayed interventions and preventative measures.

This project seeks to address these shortcomings by leveraging machine learning techniques, specifically Decision Trees and Random Forests, which are renowned for their effectiveness in handling large datasets with complex variable interactions. These models are capable of recognizing nonlinear patterns that traditional statistical approaches might overlook, offering a more nuanced understanding of stroke risk factors. The goal is to develop a more accurate and reliable stroke prediction model that can be readily implemented in clinical settings to enhance preventative healthcare strategies.

## Drop Unused Column

**Dropping the 'id' Column** <br>
- **Purpose of Dropping**: The 'id' column was dropped because it serves as a unique identifier for each row and does not contribute to the predictive modeling. Unique identifiers are typically irrelevant to the outcome of machine learning models and can actually skew the results if included, as they do not hold any intrinsic predictive value.
- **Impact on the Model**: By removing the 'id' column, the model can focus on features that genuinely affect stroke prediction, enhancing both the model's accuracy and generalizability. Keeping 'id' might lead to overfitting, where the model performs well on training data but poorly on unseen data.

```python
 #Drop the 'id' column because it's unused
 df = df.drop("id",axis=1)
```

## Filter Unused Label

**Dropping 'Other' from 'gender' Label** <br>
- **Purpose of Dropping**: The 'gender' category 'Other' was filtered out due to its minimal representation in the dataset, which is insufficient to contribute meaningfully to the model's training process. Machine learning models require a decent amount of data for each category to learn effectively; very small sample sizes can lead to biased or unstable predictive performance.
- **Impact on the Model**: Removing the 'Other' category helps in simplifying the model, ensuring that it only trains on well-represented categories (e.g., Male, Female). This focus increases the model's reliability and accuracy in predictions by preventing it from learning from noise or rare occurrences that do not generalize well across larger, more diverse datasets.

```python
# Filter the 'other' label from the gender column because it's unused
df = df[df['gender'] != 'Other']
```

### Age vs Stroke

```python
# Create a figure with two subplots, side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Create a boxplot on the first subplot (ax1)
# This shows the distribution of age for each category of stroke (0 and 1)
# The 'x' parameter is the categorical variable (stroke)
# The 'y' parameter is the numerical variable (age)
# 'data' specifies the DataFrame to use
sns.boxplot(x='stroke', y='age', data=df, ax=ax1)
ax1.set_title('Stroke vs Age')  # Set the title for the first subplot

# Create a bar plot on the second subplot (ax2)
# This shows the percentage of stroke occurrence for each age group
# First, group the data by age bins and calculate the mean of 'stroke' for each group
age_stroke_pct = df.groupby(pd.cut(df['age'], bins=np.arange(0, 100, 10)))['stroke'].mean()
# Create the bar plot
sns.barplot(x=age_stroke_pct.index.astype(str), y=age_stroke_pct.values, ax=ax2)
ax2.set_title('Percentage of Stroke by Age Group')  # Set the title for the second subplot
ax2.set_xlabel('Age Group')  # Set the x-axis label
ax2.set_ylabel('Percentage')  # Set the y-axis label

plt.tight_layout()  # Adjust the padding between and around subplots
plt.show()  # Display the plot
```



1. **Stroke vs. Age (Boxplot):** On the left side, we see a boxplot comparing the distribution of age for individuals with and without a stroke. The key observations are:
   - The median age of individuals who have had a stroke (value 1) is significantly higher than that of those who have not (value 0).
   - The interquartile range (IQR), which represents the middle 50% of the age data for those who have had a stroke, is also higher, indicating that strokes are more common in an older population.
   - There are outliers in the non-stroke group indicating younger individuals with outlier age values, but these are exceptions.

2. **Percentage of Stroke by Age Group (Bar Chart):** On the right side, there is a bar chart showing the percentage of individuals who have had a stroke within various age groups. The key points are:
   - The percentage of individuals who have had a stroke increases with age.
   - The lowest age groups (0-10, 10-20, 20-30) have the lowest percentages, which increase notably from the 40-50 age group and upwards.

```python
# Define a function to calculate the correlation ratio between categorical and continuous variables
# This function is used to measure the strength of the association between categorical variables and continuous ones
def correlation_ratio(categories, measurements):
    # Convert categorical data to numerical values
    fcat, _ = pd.factorize(categories)  # `factorize` assigns unique numbers to each unique category
    cat_num = np.max(fcat) + 1  # Determine the number of unique categories

    # Initialize arrays to hold the average measurement per category and the count of observations per category
    y_avg_array = np.zeros(cat_num)
    n_array = np.zeros(cat_num)

    # Calculate the average measurement and count for each category
    for i in range(0, cat_num):
        cat_measures = measurements[fcat == i]  # Select measurements belonging to the current category
        n_array[i] = len(cat_measures)  # Count the number of measurements in the current category
        y_avg_array[i] = np.average(cat_measures)  # Calculate the average measurement for the current category

    # Calculate the overall average measurement across all categories
    y_total_avg = np.sum(np.multiply(y_avg_array, n_array)) / np.sum(n_array)

    # Calculate the numerator for the correlation ratio (between-category variance)
    numerator = np.sum(np.multiply(n_array, np.power(np.subtract(y_avg_array, y_total_avg), 2)))

    # Calculate the denominator for the correlation ratio (total variance in the measurements)
    denominator = np.sum(np.power(np.subtract(measurements, y_total_avg), 2))

    # Calculate the correlation ratio (eta)
    if numerator == 0:
        eta = 0.0  # If there's no variance between categories, set eta to 0
    else:
        eta = np.sqrt(numerator / denominator)  # Ratio of between-category variance to total variance

    return eta  # Return the correlation ratio


# Select categorical columns from the DataFrame
categorical_cols = df.select_dtypes(include=['int']).columns  # Assuming categorical data is represented as integers

# Initialize a DataFrame for the correlation matrix between categorical variables
corr_matrix_categorical = pd.DataFrame(index=categorical_cols, columns=categorical_cols, dtype=float)

# Calculate the correlation ratio for each pair of categorical columns
for col1 in categorical_cols:  # Loop over each categorical column
    for col2 in categorical_cols:  # Nested loop for pairwise comparisons
        corr_matrix_categorical.loc[col1, col2] = correlation_ratio(df[col1], df[col2])  # Store the correlation ratio in the matrix

# Plot a heatmap to visualize the correlation matrix for categorical variables
plt.figure(figsize=(10, 8))  # Set the size of the plot
plt.title('Correlation Heatmap for Categorical Variables')  # Add a title to the plot
sns.heatmap(corr_matrix_categorical, annot=True, cmap='coolwarm', vmin=-1, vmax=1)  # Plot the heatmap
plt.show()  # Display the plot
```

In this case, the `StandardScaler` is employed to normalize the numerical columns in the dataset.

The `StandardScaler` adjusts the values in each numerical column so that they have a mean of zero and a standard deviation of one. Fitting the scaler involves computing the mean and standard deviation for each column. The same scaler is then used to transform both the training data and the testing data. This ensures that the testing data is scaled based on the parameters (mean and standard deviation) from the training data, which is critical for maintaining the validity of the model's evaluation.
- **Application to Data**: The transformation is applied only to the numerical columns specified by `numerical_cols`, adjusting these columns in both `X_train_scaled` and `X_test_scaled` to have standardized scaling.
